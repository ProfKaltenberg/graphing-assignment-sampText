```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
path='C:/Users/kanik/OneDrive/Documents/ECO 590/Class-Notes/'
league=pd.read_csv(path+'league.csv')
```

```python
league
```

## #1

```python
top = league[league['lane']=='top lane']
jgl = league[league['lane']=='jungle lane']
mid = league[league['lane']=='middle lane']
bot = league[league['lane']=='bottom lane']
sup = league[league['lane']=='support lane']

opacity=0.1

plt.plot('PBI Index', 'wr', data=top,linestyle='none', marker='.', color = 'r', label='top')
plt.plot('PBI Index', 'wr', data=jgl,linestyle='none', marker='.', color = 'g', label='jungle')
plt.plot('PBI Index', 'wr', data=mid,linestyle='none', marker='.', color = 'b', label='mid')
plt.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'y', label='bot')
plt.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'm', label='support')

plt.ylabel('Win Rate', fontweight='bold')
plt.xlabel('Pick Ban Influence', fontweight='bold')
plt.title("WR vs. PBI")

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

Pick Ban Influence (PBI) is measured by lolayltics to capture how "contested" a character choice is in the draft stage of League of Legends. The higher this value, the more the playerbase values a certain pick. The win rate is a measurement of how often people who choose a character wins. From this scatter plot, we can see a positive correlation between the characters that win the most and the characters that are contested the most.

## #2

```python
plt.hist(league['Pick Rate'], bins=30)

plt.xlabel('Pick Rate', fontweight='bold')
plt.title("Histogram of Pick Rate")
```

Pick rate is the likelihood of a specific character being chosen in a given game. Most characters have a pick rate around 0-7.5% with a few exceptions being far above every other choice.

## #3

```python
ax1 = plt.subplot(2, 2,1)
ax2 = plt.subplot(2, 2,3)

top = league[league['lane']=='top lane']
jgl = league[league['lane']=='jungle lane']
mid = league[league['lane']=='middle lane']
bot = league[league['lane']=='bottom lane']
sup = league[league['lane']=='support lane']

ax1.plot('PBI Index', 'wr', data=top,linestyle='none', marker='.', color = 'r', label='top')
ax1.plot('PBI Index', 'wr', data=jgl,linestyle='none', marker='.', color = 'g', label='jungle')
ax1.plot('PBI Index', 'wr', data=mid,linestyle='none', marker='.', color = 'b', label='mid')
ax1.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'y', label='bot')
ax1.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'm', label='support')

ax1.legend(loc='upper right', bbox_to_anchor=(1.5, 1))

ax2.hist(league['PBI Index'], bins=30)

plt.show
```

## #4

```python
sns.set_style("white")

lane = league[(league['lane']=='top lane')| (league['lane']=='jungle lane')
|(league['lane']=='middle lane') |(league['lane']=='bottom lane') | (league['lane']=='support lane') ]

ax = sns.boxplot(x='lane',y='wr',data=league, hue='lane')

ax.get_legend()

ax.set_xlabel("Lane", fontweight='bold')
ax.set_ylabel("Win Rate", fontweight='bold')
ax.set_xticks([0,1,2,3,4])
ax.set(xticklabels=['Top', 'Jungle', 'Middle','Bottom','Support'])
```

In League of Legends, there are 5 roles that players choose before entering a game. Each role has a different job and faces different dynamics. For the most part, each role is relatively balanced around the 50% win rate. However support appears to have the highest win rate on average compared to other roles while jungle is the worst performing role. Middle lane also appears to have the widest range of win rate which makes sense because it also the most popular one.
