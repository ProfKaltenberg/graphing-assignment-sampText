```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
league=pd.read_csv(r"C:\Users\ong92\Documents\GitHub\ECO 590\Class-Notes\league.csv")
```

```python
league
```

```python
top = league[league['lane']=='top lane']
jgl = league[league['lane']=='jungle lane']
mid = league[league['lane']=='middle lane']
bot = league[league['lane']=='bottom lane']
sup = league[league['lane']=='support lane']

plt.plot('PBI Index', 'wr', data=top,linestyle='none', marker='.', color = 'r', label='top')
plt.plot('PBI Index', 'wr', data=jgl,linestyle='none', marker='.', color = 'g', label='jungle')
plt.plot('PBI Index', 'wr', data=mid,linestyle='none', marker='.', color = 'b', label='mid')
plt.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'y', label='bot')
plt.plot('PBI Index', 'wr', data=bot,linestyle='none', marker='.', color = 'm', label='support')

plt.ylabel('Win Rate', fontweight='bold')
plt.xlabel('Pick Ban Influewrnce', fontweight='bold')
plt.title("WR vs. PBI")
 
# Show graphic

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

```python
plt.hist(league['PBI Index'], bins=30)

plt.xlabel('Pick Ban Influence', fontweight='bold')
plt.title("Histogram of PBI")
```

```python
sns.set_style("white")

lane = league[(league['lane']=='top lane')| (league['lane']=='jungle lane')
|(league['lane']=='middle lane') |(league['lane']=='bottom lane') | (league['lane']=='support lane') ]

ax = sns.boxplot(x='lane',y='PBI Index',data=league, hue='lane')


ax.get_legend()

ax.set_xticks([0,1,2,3,4])
ax.set(xticklabels=['top', 'jungle', 'mid','bot','support'])
```
